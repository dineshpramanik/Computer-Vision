```python
import cv2
import os
import numpy as np
import imutils
```

```python
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
from imutils.video import VideoStream
```

```python
def detect_and_predict_mask(frame, faceNet, maskNet):
    #grab the dimensions of the frame and construct a blob from it
    (h, w)= frame.shape[:2]
    blob = cv2.dnn.blobFromImage(frame, 1.0, (224,224), (104.0, 177.0, 123.0))
    
    #pass the blob through the network and obtain the face detection
    faceNet.setInput(blob)
    detection = faceNet.forward()
    print(detection.shape)
    
    #initialize list of faces, their corresponding locations, and list of predictions 
    faces = []
    locs = []
    preds = []
    
    #loop over the detection
    for i in range(0, detection.shape[2]):
        #extract the confidence (i.e probability) associated with detection
        confidence = detection[0,0,i,2]
        
        if confidence > 0.5:
            #compute the x,y co-ordinates of the bounding box
            box = detection[0,0,i,3:7] * np.array([w,h,w,h])
            (X_start, Y_start, X_end, Y_end) = box.astype('int')
            
            #ensure the vounding box fall within the dimension of the frame
            (X_start, Y_start) = (max(0, X_start), max(0,Y_start))
            (X_end, Y_end) = (min(w-1, X_end), min(h-1, Y_end))
            
            #extract the face ROI, convert it from BGR to RGB channel, resize it to 224,224 and preprocess it
            face = frame[Y_start:Y_end, X_start:X_end]
            face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
            face = cv2.resize(face, (224,224))
            face = img_to_array(face)
            face = preprocess_input(face)
            
            faces.append(face)
            locs.append((X_start, Y_start, X_end, Y_end))
            
    #only make prediction if atleast once face was detected
    if len(faces) > 0:
        faces = np.array(faces, dtype='float32')
        preds = maskNet.predict(faces, batch_size=32)
        
    return (locs, preds)
```

```python
#load the face detector model
prototxtpath = r'C:\Users\Dinesh\Desktop\Object Detection using Tensorflow\Face Mak Detection\face detector\deploy.prototxt'
weightpath = r'C:\Users\Dinesh\Desktop\Object Detection using Tensorflow\Face Mak Detection\face detector\res10_300x300_ssd_iter_140000.caffemodel'

faceNet = cv2.dnn.readNet(prototxtpath, weightpath)
```

```python
#load the face mask detector model
maskNet = load_model(r'C:\Users\Dinesh\Desktop\Object Detection using Tensorflow\Face Mak Detection\MobileNet_V2.model')
```

```python
#Initialize the Video stream
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    frame = imutils.resize(frame)
    (locs, preds) = detect_and_predict_mask(frame, faceNet, maskNet)
    
    for (box, pred) in zip(locs, preds):
        (X_start, Y_start, X_end, Y_end) = box
        (mask, without_mask) = pred
        
        #determine the class label and color
        label='Mask detected' if mask > without_mask else 'No Mask'
        color=(0,255,0) if label=='Mask detected' else (0,0,255)
        
        #Display the label and Bounding box
        cv2.putText(frame, label, (X_start, Y_start-10), cv2.FONT_HERSHEY_DUPLEX, 1, color, 2)
        cv2.rectangle(frame, (X_start, Y_start), (X_end, Y_end), color, 2) 
    
    cv2.imshow('Video', frame)
    
    if cv2.waitKey(1) == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
    
```
